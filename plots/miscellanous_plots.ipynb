```python
def lognormal(temperature, m, size=(1,),L=1):
    # temperature = temperature * (m**2)
    return np.random.lognormal(L*np.log(m), np.sqrt(L/(2*temperature)), size=size)
    # return np.random.lognormal(-L/(4*temperature)-L*np.log(m), np.sqrt(L/(2*temperature)), size=size)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import lognorm

def plot_lognormal_pdf(mu, sigma, x_range=(0, 1), num_points=1000):
    # Generate values for x-axis
    x = np.linspace(x_range[0], x_range[1], num_points)

    # Calculate PDF using lognorm.pdf() function
    pdf = lognorm.pdf(x, s=sigma, scale=np.exp(mu))

    # Plot the PDF
    plt.plot(x, pdf, label=f'mu={mu}, sigma={sigma}')
    plt.title('Lognormal PDF')
    plt.xlabel('x')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True)

# Example usage:
L=1; m=10; temperature=0.1
mu = -L/(4*temperature)-L*np.log(m)
# mu=-np.log(m)
sigma = np.sqrt(L/(2*temperature))
plot_lognormal_pdf(mu, sigma)
# mu = -L/(4*temperature)-L*np.log(m)
mu=-np.log(m)
sigma = np.sqrt(L/(2*temperature))
plot_lognormal_pdf(mu, sigma)
plt.show()
```

```python
def lognormal(temperature, m, size=(1,)):
    # temperature = temperature * (m**2)
    return np.random.lognormal(-1/(2*temperature)-np.log(m), np.sqrt(1/(2*temperature)), size=size)
```

```python
def power_law(alpha,size):
    """
    Generate samples from a power-law distribution.

    Parameters:
    alpha (float): Exponent of the power-law distribution.
    x_min (float): Minimum value for the distribution.
    size (int): Number of samples to generate.

    Returns:
    numpy.ndarray: Samples from the power-law distribution.
    """
    # Generate samples from a uniform distribution
    u = np.random.uniform(size=size)
    # Compute the samples using the inverse transform method
    return ((1 - u)) ** (-1 / (alpha - 1))
```
